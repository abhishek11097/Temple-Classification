```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import cv2
import glob
import matplotlib.pyplot as plt

APP_FOLDER="/content/train/"  #folder with entire data


totalFiles = 0 # total files
df = []
for dirpath, dirnames, filenames in os.walk(APP_FOLDER):
    counter = len(filenames)
    totalFiles += counter

    countryName = dirpath.split('/')[-1]
    if countryName != "":
        df.append([countryName ,counter])


df = pd.DataFrame(df,columns=['Country','Count'])
df.plot(x ='Country', y='Count', kind = 'bar', color = "blue")
f = plt.figure()
f.set_figwidth(5)
f.set_figheight(1)
plt.show()
```
